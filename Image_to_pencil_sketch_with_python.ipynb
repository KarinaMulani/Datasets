**Importing libraries**

```python
from google.colab.patches import cv2_imshow
import cv2
```

**Loading the image**

```python
img = cv2.imread("/content/photography .jpg")
```

```python
cv2_imshow(img)
```

**Converting an image into grey_scale image**

```python
img_grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
```

**Inverting the image**

```python
img_invert = cv2.bitwise_not(img_grey)
cv2_imshow(img_invert)
```

**Smoothing the image**

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21), sigmaX=0,sigmaY=0)
cv2_imshow(img_smoothing)
```

**Obtaining the final sketch**

```python
def dodgeV2(x,y):return cv2.divide(x,255-y,scale=256)
```

**Final output**

```python
final_img = dodgeV2(img_grey,img_smoothing)
cv2_imshow(final_img)
```
