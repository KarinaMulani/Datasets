### Task 2: Image to pencil sketch with python

**Importing libraries**

```python
from google.colab.patches import cv2_imshow
import cv2
```

**Loading the image**

```python
img = cv2.imread("/content/Tiger.jpg")
cv2_imshow(img)
```

```python
img_resize = cv2.resize(img,(300,250))
cv2_imshow(img_resize)
```

**Converting an image into gray_scale image**

```python
img_grey = cv2.cvtColor(img_resize,cv2.COLOR_BGR2GRAY)

cv2_imshow(img_grey)
```

**Inverting the image**

```python
img_invert=cv2.bitwise_not(img_grey)
cv2_imshow(img_grey)
```

```python
cv2_imshow(img_invert)
```

**Smoothing the image**

```python
img_smoothing = cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0)
cv2_imshow(img_smoothing)
```
